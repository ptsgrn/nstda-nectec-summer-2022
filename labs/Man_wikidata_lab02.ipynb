```python
import pythainlp as nlp
from pprint import pprint
import re
import pandas as pd
from pythainlp.tag import NER
```

```python
df = pd.read_csv("text-simple.csv")
ner = ""
ner_results = []
for text in df['text'].values:
  # normalized
  text = nlp.util.normalize(text)
  iner = NER("thainer")
  ner = iner.tag(text,tag=True)
  ner_results.append(ner)
```

```python
ALL_ENTITY = [
  "DATE",
  "TIME",
  "EMAIL",
  "LEN",
  "LOCATION",
  "ORGANIZATION",
  "PERSON",
  "PHONE",
  "URL",
  "ZIP",
  "MONEY",
  "LAW",
]

for ner in ner_results:
  print('\n---- [', end='')
  print(ner[:5],end='')
  print('] ----')
  for entity in ALL_ENTITY:
    try:
      p = re.compile(f'<{entity}>(.+?)</{entity}>')
      result = p.findall(ner)
    except AttributeError:
      result = ''
    print(entity, ':', end=' ')
    pprint(result)
```
